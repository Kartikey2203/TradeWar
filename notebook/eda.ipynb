```python
tariff_map = {

    # ===================== Technology =====================
    **{(y, "Telecommunication Equipment"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
    **{(y, "Computers & Data Processing"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
    **{(y, "Integrated Circuits (Semiconductors)"): (0 if y == 2017 else (50 if y >= 2024 else 25)) for y in range(2017, 2026)},
    **{(y, "Semiconductors"): (0 if y == 2017 else (50 if y >= 2024 else 25)) for y in range(2017, 2026)},
    **{(y, "Monitors & Television Receivers"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
    **{(y, "Audio Equipment (Speakers & Microphones)"): (0 if y == 2017 else 25) for y in range(2017, 2026)},

    # ===================== Automobile =====================
    **{(y, "Passenger Vehicles"): (2.5 if y < 2024 else 100) for y in range(2017, 2026)},
    **{(y, "Vehicle Parts & Accessories"): (2.5 if y == 2017 else (25 if y < 2024 else 25)) for y in range(2017, 2026)},

    # ===================== Energy =====================
    **{(y, "Electric Batteries"): (0 if y < 2024 else 25) for y in range(2017, 2026)},
    **{(y, "Crude Petroleum Oil"): 0 for y in range(2017, 2026)},
    **{(y, "Petroleum Gases (LNG/LPG)"): (0 if y == 2017 else 25) for y in range(2017, 2026)},

    # ===================== Agriculture =====================
    **{(y, "Soybeans"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
    **{(y, "Maize (Corn)"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
    **{(y, "Grain Sorghum"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
    **{(y, "Pork"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
    **{(y, "Wheat"): (0 if y == 2017 else 25) for y in range(2017, 2026)},

    # ===================== Healthcare =====================
    **{(y, "Medical & Surgical Instruments"): (0 if y == 2017 else 25) for y in range(2017, 2026)},

    # ===================== Critical Minerals =====================
    **{(y, "Rare-Earth Compounds"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
    **{(y, "Alkali & Alkaline-Earth Metals"): (0 if y == 2017 else 25) for y in range(2017, 2026)},

    # ===================== Industrial =====================
    **{(y, "Electromagnets & Magnetic Devices"): (0 if y == 2017 else 25) for y in range(2017, 2026)},
}

trade_total_clean["Tariff_Rate (%)"] = trade_total_clean.apply(
    lambda row: tariff_map.get((row["Year"], row["Category"]), 0),
    axis=1
)
```

```python
sector_year = (
    trade_total_clean
    .groupby(["Year", "Sector"])["Trade_Value_USD"]
    .sum()
    .reset_index()
)

sector_year.head()
```

```python
sector_pivot = sector_year.pivot(
    index="Sector",
    columns="Year",
    values="Trade_Value_USD"
)

sector_pivot
```

```python
sector_pivot["Impact (%)"] = (
    (sector_pivot["Year"] - sector_pivot[2017])
    / sector_pivot[2017]
) * 100
```
